#### #How to generate numpy arrays using random module:
##### syntax:
import numpy as np
np.random.rand()  -> It generates numbers uniformly between 0 and 1
np.random.randn() -> It generate normally distributed numbers between 0 and 1
np.random.ranint() -> It generate uniformly distributed numbers between 0 and given number.

```python
## package-module-methods/functions
import numpy as np
print(np.random.randn())
print(np.random.rand(3,3))
print(np.random.rand()+100*50)
print(np.random.randn())
print(np.random.randn(3,2))
```

```python
print(np.random.randint(30))
print(np.random.randint(50,size=(4,5)))
```

```python
##numpy operations on numpy arrays:
import numpy as np
a1=np.array([1,23,50,90,"priya","python"])
print(a1[0])
print(a1[4])
print(a1[5])
print(a1[0:5])
```

```python
#Indexing on multi-dimensional array
a1=np.array([[50,60,"priya"],[100,500,"python"],[90.3,600,"Data Science"]])
print(a1[1][2])
print(a1[1,2])
print(a1[2,2])
print(a1[:,2])
print(a1[0:3,2])
```

```python
##Stacking on Numpy arrays(hstacking and vstacking)
import numpy as np
a1=np.array([[1,2,3],[4,5,6]])
a2=np.array([[7,8,9],[10,11,12]])
print(a1,a2,sep="\n")
a3=np.hstack((a1,a2)) # Stack arrays in sequence horizontally (column wise). This is equivalent to concatenation along the second axis, except for 1-D
print(a3)
a4=np.vstack((a1,a2)) #Stack arrays in sequence vertically (row wise). This is equivalent to concatenation along the first axis after 1-D arrays of shape `(N,)` have been reshaped to `(1,N)`. Rebuilds arrays divided by `vsplit`
print(a4)
```

```python
import numpy as np
a1=np.array([[1,2,3],[4,5,6]])
print("Original array:",a1)
aT=a1.transpose()
print("Transpose array:",aT)
```

###### Matplotlib in Python:
    It is used for 2d graphics in python programming to represent the complex data.
###### Types of plots:
    1. Plots
    2. Pie charts
    3. Bar graphs
    4. Area graphs
    5. Scatter plots
    6. Histograms

```python
# How to generate normal plot:
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")  ##for colouring the plot
x=[5,6,8]
y=[12,6,9]
x1=[1,2,3]
y1=[4,5,1]
plt.plot(x,y,"g",label="1st line",linewidth=3)
plt.plot(x1,y1,"c",label="2nd line",linewidth=3)
plt.title("This is about Normal Plot")
plt.xlabel("X data points")
plt.ylabel("Y data points")
plt.grid(True,color="k")
#to view the plot
plt.show()
```
